## Feature Scaling 

    Feature scaling is a technique to standardize the independent features present in the data in a fixed range.

### Why we do Feature Scaling?
    -> If we calculate the distance b/w salary and Age the Salary will dominate in data and the algorithm cannot perform well.

### Types of feature Scaling:-
    1.Standardization
    2.Normalization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('D:\Machine_learning\ML_With_Campusx\csv_files\Social_Network_Ads.csv')
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
df.sample(5)
```

```python
x=df.drop(columns='Purchased')
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)

x_train.shape,x_test.shape
```

# Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()

# fit the scalar to the train set,it will learn the parameter
scalar.fit(x_train)

#transform train and tests
scaled_x_train=scalar.transform(x_train)
scaled_x_test=scalar.transform(x_test)
```

```python
scalar.mean_
```

```python
scaled_x_test
```

```python
scaled_x_test
```

```python
scaled_x_train=pd.DataFrame(scaled_x_train,columns=x_train.columns)
scaled_x_test=pd.DataFrame(scaled_x_test,columns=x_test.columns)
```

```python
scaled_x_test
```

```python
scaled_x_train
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(scaled_x_train.describe(),1)
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2 , figsize=(12,5))

ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(scaled_x_train['Age'],scaled_x_train['EstimatedSalary'],color='red')
ax2.set_title('After Scaling')
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

#before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)

#After Scaling
ax2.set_title('After Scaling')
sns.kdeplot(scaled_x_train['Age'],ax=ax2)
sns.kdeplot(scaled_x_train['EstimatedSalary'],ax=ax2)
plt.show()
```

# Comparison of Distribution

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

#Before Scaling
ax1.set_title('Age Distribution before scaling')
sns.kdeplot(x_train['Age'],ax=ax1)

#After Scaling
ax2.set_title('Age Distribution after scaling')
sns.kdeplot(scaled_x_train['Age'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

#Before Scaling
ax1.set_title('Salary distribution before scaling')
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)

#After Scaling
ax2.set_title('Salary distribution after scaling')
sns.kdeplot(scaled_x_train['EstimatedSalary'],ax=ax2)
plt.show()
```

# Why scaling is important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
lr_scaled.fit(scaled_x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred_scaled=lr_scaled.predict(scaled_x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Actual',accuracy_score(y_test,y_pred))
print('Scaled',accuracy_score(y_test,y_pred_scaled))
```

```python
#No effect of scaling on DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt_scaled=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
dt_scaled.fit(scaled_x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred_scaled=dt.predict(scaled_x_test)
```

```python
print('Actual',accuracy_score(y_test,y_pred))
print('Scaled',accuracy_score(y_test,y_pred_scaled))
```
