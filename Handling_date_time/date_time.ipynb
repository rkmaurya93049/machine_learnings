# Working with date

```python
# Converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

### Extract Year 

```python
date['date_year']=date['date'].dt.year
date.head()
```

### Extract Month

```python
date['date_month']=date['date'].dt.month
date.head()
```

```python
date['date_month_Name']=date['date'].dt.month_name()
date.head()
```

### Extract Day

```python
date['date_day']=date['date'].dt.day
date.head()
```

```python
date['date_dow']=date['date'].dt.dayofweek
date.head()
```

```python
# days of the week name
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
date.head()
```

```python
date['dow_name']=date['date'].dt.day_name()
date.head()
```

```python
# is Weekend?
date['date_is_weekend']=np.where(date['dow_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

# Extract week of the year

```python
date['date_week']=date['date'].dt.isocalendar().week
date.head()
```

```python
date['date_quater']=date['date'].dt.quarter
date.head()
```

# Extract Semester

```python
date['semester']=np.where(date['date_quater'].isin([1,2]),1,2)
date.head()
```

# Extract time elapsed between dates

```python
import datetime

today=datetime.datetime.today()

today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
# month passed

np.round((today-date['date'])/np.timedelta64(1,'M'),0)
```
