```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python

webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
soup.find_all('h1')[0].text.strip()
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for r in soup.find_all('span',class_='companyCardWrapper__companyRatingValue'):
  print(r.text.strip())
```

```python
for r in soup.find_all('span',class_='companyCardWrapper__ActionCount'):
  print(r.text.strip())
```

```python
for i in soup.find_all('span',class_='companyCardWrapper__ratingValues'):
  print(i.text.strip())
```
